```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


df = pd.read_csv('movies.csv')

print("Setup Complete. Ready to preditct.")
```

```python
X = df[['Rating']]

y = df['Revenue']

print("Shape of X: ", X.shape)
print("Shape of y: ", y.shape)
```

```python
model = LinearRegression()

print("Brain created. It knows nothing yet.")
```

```python
model.fit(X, y)

print("Training complete.")
```

```python
print("Slope:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
# Ask the model to predict the revenue for an 8.0 rating
prediction = model.predict([[8.0]])

print(f"Predicted Revenue for an 8.0 movie: ${prediction[0]:.2f} Million")
```

```python
# 1. Draw the dots 
plt.scatter(df['Rating'], df['Revenue'], color='blue', label='Actual Movies')

# 2. Draw the line 
plt.plot(df['Rating'], model.predict(X), color='red', linewidth=3, label='AI Prediction Line')

# 3. Add labels and legend
plt.title('Movie Revenue Prediction')
plt.xlabel('Rating')
plt.ylabel('Revenue ($ Million)')
plt.legend()
plt.show()
```
