```python
import pandas as pd

df = pd.read_csv('movies.csv')

df.head()
```

```python
myfavs = df.head() 
```

```python
action_movies = df[ df['Genre'] == 'Action' ]

action_movies
```

```python
movies = df[ (df['Genre'] != 'Action') ]

movies
```

```python
import matplotlib.pyplot as plt
```

```python
x_data = myfavs['Title']
y_data = myfavs['Revenue']

colors = ['red', 'blue', 'green', 'orange', 'purple']

plt.bar(x_data, y_data, color=colors)

plt.ylabel('$')
plt.xlabel('Movie Title')
plt.title('My Favorite Movie Revenues')

plt.show()
```

```python
df.groupby('Genre')['Revenue'].mean()
```

```python
# 1. Do the Math and SAVE it
genre_avg = df.groupby('Genre')['Revenue'].mean()

# 2. Plot it
# genre_avg.index = The names (Action, Drama, etc.)
# genre_avg.values = The numbers (1000, 2187, etc.)
plt.bar(genre_avg.index, genre_avg.values, color='purple')

# 3. Add Labels
plt.title('Average Revenue by Genre')
plt.ylabel('Revenue ($ Million)')
plt.xlabel('Genre')

plt.show()
```

```python
genre_avg.to_csv('genre_report.csv')
```
