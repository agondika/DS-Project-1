```python
import pandas as pd

df = pd.read_csv('movies.csv')

df.head()
```

```python
myfavs = df.head() 
```

```python
import matplotlib.pyplot as plt
```

```python
x_data = myfavs['Title']
y_data = myfavs['Revenue_Millions']

colors = ['red', 'blue', 'green', 'orange', 'purple']

plt.bar(x_data, y_data, color=colors)

plt.ylabel('$')
plt.xlabel('Movie Title')
plt.title('My Favorite Movie Revenues')

plt.show()
```
